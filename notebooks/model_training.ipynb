```python
# Define the custom ranking for each ordinal variable
cut_categories = ['Fair', 'Good', 'Very Good','Premium','Ideal']
color_categories = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
clarity_categories = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
from sklearn.impute import SimpleImputer ## HAndling Missing Values
from sklearn.preprocessing import StandardScaler # HAndling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding
## pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# Piepline :
# pipeline is set of automated processes or tools that are connected in a sequence to
# perform a specific tasks
```

```python
num_pipeline=Pipeline(
    
    
    steps=[
        
        ('imputer',SimpleImputer()), # it handle missing values
        ('scaler', StandardScaler())  # it reduce the size of the data
    ]
    
    
)
    
```

```python
cat_pipeline=Pipeline(
    
    steps=[
        ('imputer',SimpleImputer(strategy='most_frequent')), # handle the repeated values 
        ('ordinalencoder',OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories])) # provide the ranking for the data 
    ]
    
)
```

```python
preprocessor=ColumnTransformer(
    [
        
        ('num_pipeline',num_pipeline,numerical_cols),
        ('cat_pipeline',cat_pipeline,categorical_cols)
    ]
)
```

```python

## Train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=30)
```

```python
preprocessor.fit_transform(X_train)
```

```python
preprocessor.transform(X_test)
```

```python
# accessing the column name of the features over present in the data
preprocessor.get_feature_names_out()
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
## Model Training

# importing the machien learning model
#import the perfomance matrices

from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
import numpy as np
# here i am evaluating my model
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
## Train multiple models
# for that i crate a list which have multiple models

models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet()
}
```

```python
trained_model_list=[]
model_list=[]
r2_list=[]
```

```python
list(models)
```

```python
# using it we get the values of the models
for i in range(len(list(models))):
    model=list(models.values())[i]
    print(model)
```

```python
models.keys()
```

```python
models.values()
```

```python

# here i tarain my model which are linearRegression lasso ridge elastic net
for i in range(len(list(models))):
    # here i get each model 
    model=list(models.values())[i]
    # here i train my machine learning model on the train data
    
    model.fit(X_train,y_train)

    #Make Predictions
    y_pred=model.predict(X_test)

    #this is a validation(test) score
    mae, rmse, r2_square=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')
```
